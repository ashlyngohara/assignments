```python
import agate
import agatestats
import csv
```

```python
crime_analysis = agate.Table.from_csv('campus_crime.csv')
```

```python
print(crime_analysis)
```

```python
inst_enrollment = crime_analysis.order_by('total_enr', reverse = True)
```

```python
inst_enrollment.print_table(max_rows = None)
```

```python
outliers = inst_enrollment.stdev_outliers('forcible')
```

```python
outliers.print_table(max_rows=None)
```

```python
murder_outliers = crime_analysis.order_by('murder', reverse = True)
```

```python
murder_outliers.print_table(max_rows=None)
```

Even though San Jose State University, the University of Arizona, the University of Illinois and Virginia Polytech do not have the higest enrollment numbers, they are the only ones to report murders. Could do further analysis to figure out what makes these schools unique. 

```python
forcible_outliers = crime_analysis.order_by('forcible', reverse=True)
```

```python
forcible_outliers.print_table(max_rows=None)
```

```python
arson_outliers = crime_analysis.order_by('arson', reverse = True).exclude('manslaughter').exclude('nonforce').exclude('murder').exclude('forcible').exclude('total_enr')
```

```python
arson_outliers.print_table(max_rows=None)
```

```python
burglaries = crime_analysis.order_by('burglary', reverse=True).exclude('manslaughter').exclude('robbery').exclude('forcible')
```

```python
burglaries.print_table(max_rows = None)
```

Would be interested as to why these top four universities for burgalries are what they are, as their total enrollment is not as high as others.

```python
manslaughters = crime_analysis.order_by('manslaughter', reverse=True)
```

```python
manslaughters.print_table(max_rows=None)
```

Florida State University is the only university on the list where 'manslaughter' = True. Definitely worth investigating.

```python
crimes = crime_analysis.order_by('murder')
summary = crimes.aggregate([
    ('manslaughter', agate.Count()),
    ('mean_murder', agate.Mean('murder')),
    ('mean_agg_aslt', agate.Mean('agg_aslt'))
])
```

```python
summary.print_table(max_rows=None)
```

```python
mean_burglaries = crime_analysis.aggregate(agate.Mean('burglary'))
```
